```python
# Iniciamos una lista para almacenar los datos
datos = []
num_tweets = 3000

# Recorremos el Cursor con un "for"
for tweet in tweepy.Cursor(api.search, q=('covid' or 'coronavirus') and ('vacuna' or 'vaccine'),
                           result_type='recent', tweet_mode='extended').items(num_tweets):

    # Inicializamos la variables a cadena vac√≠a
    relation = ""
    user_retweeted = ""
    user_replied = ""
    user_quoted = ""
    firstHT = ""
    
    # Recuperamos los datos
    # Retweet
    try:
        user_retweeted = tweet.retweeted_status.user.screen_name
        TweetCompleto = 'RT @' + user_retweeted + ': ' + tweet.retweeted_status.full_text.replace("\n", "")
        relation = "RT"
    except AttributeError:  # No es un Retweet
        TweetCompleto = tweet.full_text.replace("\n", "")
        
    # Reply
    if (tweet.in_reply_to_screen_name != None):
        user_replied = tweet.in_reply_to_screen_name
        relation = "reply"
    
    # Quote
    try:
        user_quoted = tweet.quoted_status.user.screen_name
        relation = "quote"
    except AttributeError: # No es una citaci√≥n
        pass
    
    # Hashtags
    try:
        firstHT = tweet.entities['hashtags'][0]
    except IndexError: # No tiene hashtags
        pass

    datos.append([tweet.id, tweet.created_at, tweet.user.screen_name, TweetCompleto, tweet.source,
                  tweet.user.location, relation, user_replied, user_retweeted, user_quoted, firstHT, tweet.lang,
                  tweet.user.created_at])
    
# Convertimos la lista en un DataFrame y nombramos las columnas
datos = pd.DataFrame(datos, columns=['id tweet', 'date', 'author', 'text', 'app', 'location', 'relation', 'user replied',
                                    'user retweeted', 'user quoted', 'first HT', 'lang', 'created_at'])
# Guardamos en el directorio en que estamos trabajando
datos.to_csv('tweetsDescargados_2.csv', sep = '\t', encoding="UTF-8")
print(datos)
```
